```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array(["A","B","C","D","E"])
y = np.array([1,2,3,4,5])
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("My-Graph")
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("My-Graph")
plt.grid()
plt.show()
```

```python
plt.plot(x,y,color = "red")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("My-Graph")
plt.grid()
plt.show()
```

```python
plt.plot(x,y,ls= ":")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("My-Graph")
plt.grid()
plt.show()

"""
plt.plot(x,y,ls= ":")

Solid	  | 'solid'   | '-'  or None
Dashed	  | 'dashed'  | '--'
Dash-Dot  |	'dashdot' | '-.'
Dotted	  | 'dotted'  | ':'
No Line	  | 'None'	  | ''   or ' '

"""

```

```python
x = np.array(["A","B","C","D","E"])
y = np.array([1,2,2,7,5])
plt.plot(x,y,linewidth = "2.5",marker = 'o', ms = 10, mec = 'g',mfc = 'r')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("My-Graph")
plt.grid()
plt.show()



# ----Markers------
# 'o'	Circle	
# '*'	Star	
# '.'	Point	
# ','	Pixel	
# 'x'	X	
# 'X'	X (filled)	
# '+'	Plus	
# 'P'	Plus (filled)	
# 's'	Square	
# 'D'	Diamond	
# 'd'	Diamond (thin)	
# 'p'	Pentagon	
# 'H'	Hexagon	
# 'h'	Hexagon	
# 'v'	Triangle Down	
# '^'	Triangle Up	
# '<'	Triangle Left	
# '>'	Triangle Right	
# '1'	Tri Down	
# '2'	Tri Up	
# '3'	Tri Left	
# '4'	Tri Right	
# '|'	Vline	
# '_'   Hline	


```
